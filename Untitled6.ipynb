```python
def build_cmapss_columns():
    """
    CMAPSS FD001 format:
      - engine_id
      - cycle
      - 3 operational settings
      - 21 sensors
    Some raw files can include trailing whitespace columns -> we'll drop NaN cols after reading.
    """
    cols = ["engine_id", "cycle", "op_altitude", "op_mach", "op_throttle"]
    cols += [f"sensor_{i}" for i in range(1, 22)]
    return cols
```

```python
def load_cmapss(train_path, test_path, rul_path):
    """
    Loads train/test with robust whitespace parsing.
    Drops any all-NaN columns produced by trailing separators.
    """
    cols = build_cmapss_columns()

    train_df = pd.read_csv(train_path, sep=r"\s+", header=None, names=cols, engine="python")
    test_df  = pd.read_csv(test_path,  sep=r"\s+", header=None, names=cols, engine="python")

    # True RUL for test engines (one value per engine)
    y_true = pd.read_csv(rul_path, header=None).values.flatten()

    # Drop empty columns if any (common in CMAPSS txt files)
    train_df.dropna(axis=1, inplace=True)
    test_df.dropna(axis=1, inplace=True)

    return train_df, test_df, y_true
```

```python
def compute_train_rul(train_df):
    """
    RUL per row = max_cycle(engine) - current_cycle.
    Vectorized and correct.
    """
    max_cycles = train_df.groupby("engine_id")["cycle"].max()
    train_df = train_df.copy()
    train_df["RUL"] = train_df["engine_id"].map(max_cycles) - train_df["cycle"]
    return train_df


def select_last_cycle_per_engine(df):
    """
    Safely select the last recorded cycle per engine_id.
    We sort by (engine_id, cycle) then take tail(1) for each engine.
    """
    last_df = (df.sort_values(["engine_id", "cycle"])
                 .groupby("engine_id", as_index=False)
                 .tail(1))
    return last_df


def rmse(y_true, y_pred):
    return float(np.sqrt(mean_squared_error(y_true, y_pred)))
```
